# Part 1. Linear Case

```python
# estimate the mean distance between two numbers from [0, 1] assuming uniform distribution of selected numbers
```

```python
import numpy as np
n = 11100
X = np.random.uniform(0,1, n)
Y = np.random.uniform(0,1, n)

mean_distance = np.mean(np.abs(X-Y))

print(mean_distance)
```

```python
#N = Z**2 * p * (1-p)/err**2
# p - is a probability of a binary event outcome

#1.96**2 * (0.5) * (1 - 0.5) / 0.01**2

# for non binary event p * (1 - p) equals to Standard Deviation:
# for uniform distribution std_dev = np.sqrt((b - a)**2/12)
std_dev = np.sqrt((1 - 0)**2/12)
print(std_dev)

1.96**2 * std_dev / 0.01**2
```

# Part 2. 2D space problem

```python
from matplotlib import pyplot as plt

# estimate the mean distance between two points in 2D space rounded with circle with radius = 1

# solve with polar coordinate
# distance formula in polar coords = np.sqrt( r1**2 + r2**2 - 2* r1 * r2 * np.cos(rho))
n = 100
r1 = np.random.uniform(0,1,n)
rho = np.random.uniform(0, 2*np.pi, n)


plt.polar(rho, r1, '.')
plt.show()
```

```python
# THE PROBLEM: we noticed that 50% of points are concentrated whithin radius <= 0.5,
# which contains only 25% of circle's area

# to fix this issue we have to transform number generation for r to make it uniform in terms of circle distribution
n = 100
r1 = np.sqrt(np.random.uniform(0,1,n))
rho = np.random.uniform(0, 2*np.pi, n)

plt.polar(rho, r1, '.')
plt.show()
```

```python
# distance calculation
n = 10000
r1 = np.sqrt(np.random.uniform(0,1,n))
r2 = np.sqrt(np.random.uniform(0,1,n))
rho1 = np.random.uniform(0, 2*np.pi, n)
rho2 = np.random.uniform(0, 2*np.pi, n)


d = np.sqrt( r1**2 + r2**2 - 2* r1 * r2 * np.cos(rho1))

np.mean(d)
```
