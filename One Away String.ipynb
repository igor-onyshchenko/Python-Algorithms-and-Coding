```python
# Implement your function below.
def is_one_away(s1, s2):

    def _diff_length(s1, s2):
        mismatch = 0
        for n, v in enumerate(s2):
            if s1[n-mismatch] != s2[n]:
                mismatch += 1
                if mismatch > 1:
                    return False

    #check the length
    length_s1 = len(s1)
    length_s2 = len(s2)
    if abs(length_s1 - length_s2) > 1:
        return False

    # case 1 - one character is replaced for another
    mismatch = 0
    if length_s1 == length_s2:
        for n,v in enumerate(s1):
            if v != s2[n]:
                mismatch += 1
                if mismatch > 1:
                    return False

    # case 2,3 - if one character is deleted/inserted
    if abs(length_s1 - length_s2) == 1:
        if length_s1 < length_s2:
            _diff_length(s1, s2)

        else:
            _diff_length(s2, s1)


    return True

```

```python
print(is_one_away("abbc", "abbad"))  # should return True
print(is_one_away("abc", "bcc"))  # should return False
```
