```python
# Implement your function below.
def most_frequent(given_list):

    max_item = None
    max_value = 0
    dict_items = {}

    for i in given_list:

        if i in dict_items.keys():
            dict_items[i] += 1

        else:
            dict_items[i] = 1

        if dict_items[i] > max_value:
            max_value = dict_items[i]
            max_item = i


    return max_item


# NOTE: The following input values will be used for testing your solution.
# most_frequent(list1) should return 1
list1 = [1, 3, 1, 3, 2, 1]
print(most_frequent(list1), 1)
# most_frequent(list2) should return 3
list2 = [3, 3, 1, 3, 2, 1]
print(most_frequent(list2), 3)
# most_frequent(list3) should return None
list3 = []
print(most_frequent(list3), 'None')
# most_frequent(list4) should return 0
list4 = [0]
print(most_frequent(list4), 0)
# most_frequent(list5) should return -1
list5 = [0, -1, 10, 10, -1, 10, -1, -1, -1, 1]
print(most_frequent(list5), -1)
```
